# Hello world of data science example

Import the `pandas` package

```python
import pandas as pd
```

Obtain the Iris data set

```python
url = "https://raw.githubusercontent.com/jbrownlee/Datasets/master/iris.csv"
names = ['sepal length [cm]', 'sepal width [cm]', 'petal length [cm]', 'petal width [cm]', 'iris type']
df = pd.read_csv(url, names=names)
```

Have a quick look at the data

```python
df
```

Plot the data with features $(x_1, x_2)\in\mathbb R^2$ where:
* $x_1$ denotes the sepal length in cm and
* $x_2$ denotes the sepal width in cm.

```python

df_setosa = df[df['iris type'] == 'Iris-setosa']
ax = df_setosa.plot.scatter(x='sepal length [cm]', y='sepal width [cm]', c='red', label='Iris-setosa')

df_setosa = df[df['iris type'] == 'Iris-versicolor']
ax = df_setosa.plot.scatter(x='sepal length [cm]', y='sepal width [cm]', c='green', label='Iris-versicolor', ax=ax)

df_setosa = df[df['iris type'] == 'Iris-virginica']
df_setosa.plot.scatter(x='sepal length [cm]', y='sepal width [cm]', c='blue', label='Iris-virginica', ax=ax)
```
